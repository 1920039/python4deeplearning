# 여러 곳을 한꺼번에 수정하기: 멀티 커서(multi-cursor)

아래 코드의 'not_selling'이라는 딕셔너리 이름을 모두 'nonbiz_item'으로 바꾸려 한다. not_selling은 코드에서 세 군데에 나온다. 이를 한꺼번에 바꾸는 방법이 없을까?

```python
stock = {'tv':5, 'monitor':3, 'oven': 8}
order = {'tv':2, 'hairdrier':3, 'monitor':2}

in_stock = {}
not_selling = {}
for key, value in order.items():
    try:
        in_stock[key] = stock[key] - value
    except:
        not_selling[key] = value
print(in_stock)
print(not_selling)
```

① 첫번째 not_selling 바로 앞에서 클릭하여 그곳에 커서를 위치시키고, 키보드의 알트(Alt) 키를 누른 상태에서 바꾸고 싶은 다른 곳들을 하나하나 클릭하면 커서가 세 개 생겨 깜박인다.

image.png

② 이 상태에서 지우기(delete)를 하고 타이핑을 하면 세 군데에 모두 적용된다.

```python
stock = {'tv':5, 'monitor':3, 'oven': 8}
order = {'tv':2, 'hairdrier':3, 'monitor':2}

in_stock = {}
nonbiz_item = {}
for key, value in order.items():
    try:
        in_stock[key] = stock[key] - value
    except:
        nonbiz_item[key] = value
print(in_stock)
print(nonbiz_item)
```

여기서는 수정하고 싶은 단어가 세 군데만 있지만 더 많을 경우에는 더 유용하다. 만약 대량의 '바꾸기'를 할 때는 메뉴의 '찾기 및 바꾸기(Ctrl + H)'를 이용하는 방법이 더 편하겠지만, 바꿀 내용이 몇 개 정도일 때에는 이 방법이 더 직관적이다.

# Tab키는 문단에도 먹는다: Tab과 Shift-tab

다음과 같은 코드를 작성하였는데, 이를 함수에 넣고 싶어졌다. def 함수명( ): 을 쓰고 기존 코드를 4칸씩 오른쪽으로 옮겨서 들여쓰기를 해야 하는데, 하나하나 움직이지 않고 한꺼번에 해결하려면 어떻게 할까?

```python
in_stock = {}
not_selling = {}
for key, value in order.items():
    try:
        in_stock[key] = stock[key] - value
    except:
        not_selling[key] = value
print(in_stock)
print(not_selling)
```

마우스로 대상 영역을 선택한 뒤 키보드의 tab(탭) 키를 눌러주면 한꺼번에 이동한다.

image.png

```python
def check_stock():
    in_stock = {}
    not_selling = {}
    for key, value in order.items():
        try:
            in_stock[key] = stock[key] - value
        except:
            not_selling[key] = value
    print(in_stock)
    print(not_selling)
    return
```

만약 '설정'에서 '들여쓰기 너비(공백 개수)'를 2로 바꾸어 둔 상태라면 tab을 눌렀을 때 2칸만 움직인다.

image.png

위의 반대 작업, 즉 들여쓰기를 없애고 싶으면 tab 대신 shift-tab을 누르면 된다
